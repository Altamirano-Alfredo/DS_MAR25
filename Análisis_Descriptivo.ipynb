```python
import seaborn as sns
import matplotlib.pyplot as plt
#matplotlib inline
# Definir datos de dibujo
x = [1, 2, 3, 4, 5, 6, 7, 8]
y_bar = [5, 9, 19, 32, 28, 15, 8, 4]
y_line = [2, 9, 5, 7, 8, 9, 8, 10, 6, 7]
sns.set(context='poster',
        style='white',
        palette='dark',
        font='sans-serif',
        font_scale=0.55, color_codes='r', rc=None)
plt.bar(x, y_line)
plt.xlabel('Límite de clase')
plt.ylabel('Volumen')
plt.title('Total de productos que han experimentado variaciones')
plt.legend()
```
